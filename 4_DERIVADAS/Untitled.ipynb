### Método del Intervalo Cerrado
Encuentre el máximo absoluto y el mínimo absoluto de la función dada usando el método del intervalo cerrado.

$f(x) = 3x^2 - 12x + 5$ en $[0,3]$

Comenzamos por localizar la derivada de la función; despues encontrar cualquier número crítico en el intervalo dado determinando donde la derivada es igual a cero o indefinida. Nota: al encontrar números críticos, estarás localizando posibles puntos de **crestas** o **valles** en la gráfica.

La derivada de $f(x) = 3x^2 - 12x + 5$ es

```python
from sympy import Symbol, diff, solve
x = Symbol('x')
fx = 3*x**2-12*x+5 #Función original
dfx = diff(fx,x)
dfx
```

Igualando la derivada a 0 y resolviendo la ecuación tenemos el valor crítico $x = 2$

```python
sol = solve(dfx,x)
sol[0]  
```

Sustituimos los extremos del intervalo $[0,3]$ en la función original y elegimos el mayo y menor:

```python
f_0 = fx.subs(x,0)
f_2 = fx.subs(x,2)
f_3 = fx.subs(x,3)
print("f(0) = ",f_0, "\nf(2) = ", f_2, "\nf(3) = ", f_3)
```

Por lo tanto, el máximo absoluto es **5**, y el mínimo absoluto es **-7**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-0.1,3.1,100) # Generamos un arreglo desde 0.1 hasta 3.1 con 100 valores
fx = 3*x**2-12*x+5 # Definimos la variable
plt.plot(x,fx,label = 'f(x)') # Gráficamos la función f(x)

plt.legend(loc = 'upper center') # Ubicamos la etiqueta de la función arriba al centro

# Nombre de los ejes
plt.xlabel('x', fontsize = 15) 
plt.ylabel('f(x)', fontsize = 15)

# Graficamos los puntos, limites y punto crítico
plt.grid( color='gray', alpha = 0.6, linestyle = 'dotted', lw = 1.5)
plt.scatter([0,2,3],[f_0,f_2,f_3],color = 'r', marker = 'v')
plt.show()
```

### Encontrando intervalos de incremento y decremento

Encuentra los intervalos de incremento y decremento, si los hay, para la función:
$$f(x) = 2 \cos x – \cos 2x \ \ en \ \ 0 \le x \le 2π$$


Iniciamos por declarar la función y encontrar la derivada.

```python
from sympy import cos
x = Symbol('x')
fx = 2*cos(x)- cos(2*x)
fx
```

```python
dfx = diff(fx,x) #cálculo de la derivada
dfx
```

Ahora encontramos los valores críticos resolviendo la ecuación igualada a cero para x.


```python
sol = solve(dfx,x)
len(sol) #verificamos el número de soluciones encontradas
```
